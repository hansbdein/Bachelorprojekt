```python
#import warnings
#warnings.filterwarnings('ignore')
import numpy
from IPython.display import Image
from galpy.util import bovy_plot
bovy_plot.bovy_print(axes_labelsize=17.,text_fontsize=12.,
                     xtick_labelsize=15.,ytick_labelsize=15.)
%pylab inline






from galpy.potential import SphericalShellPotential, RingPotential
figsize(8,6)
sp= SphericalShellPotential(amp=1.,a=1.5)
rp= RingPotential(amp=1.,a=1.5)
rs= numpy.linspace(0.1,7.,101)
line_shell= bovy_plot.bovy_plot(rs,[sp(r,0.1*r) for r in rs],
                                yrange=[-1.,0.]#,
                                #xlabel=r'$r$',
                                #ylabel=r'$\Phi$',
                                #label=r'$\mathrm{Shell}$'
                                )
line_ring= bovy_plot.bovy_plot(rs,[rp(r,0.1*r) for r in rs],overplot=True#,
                               #label=r'$\mathrm{Ring}$'
                               )
legend(handles=[line_shell[0],line_ring[0]],
       fontsize=18.,loc='lower right',frameon=False);
```
